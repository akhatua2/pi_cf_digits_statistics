# Single Digit Statistics and Z-Scores

```python
import numpy as np
from scipy import stats
import math
from collections import defaultdict, Counter
# import matplotlib.pyplot as plt
from tqdm import tqdm
# from scipy.signal import savgol_filter
import pandas as pd
import os  
import time
```

```python
pi_digits = np.load("iteration1_int.npz")
```

```python
pi_cf_digits = list(pi_digits['arr_0'])[1:10000001] # use 10M and blocksize 100M
```

```python
def zscore(data, blocksize=100000, d = 1):

    number_of_blocks = len(data)/blocksize
    
    groups = list(np.array_split(data, number_of_blocks))
    k = []
    for each_list in groups:
        k.append(list(each_list).count(d))
    #print(k)
    P = math.log2(1+1/(d*(d+2)))
    z_list = []
    for i in range(int(number_of_blocks)):
        z_list.append((k[i]-blocksize*P)/(math.sqrt(blocksize*P*(1-P))))
          
    return z_list
```

# Kolmogorov-Smirnov Test for z_score

```python
pi_cf_digits = list(pi_digits['arr_0'])[1:10000001] 
```

```python
def ks_zscore(data, blocksize) :
    number_of_blocks = len(data)/blocksize
    
    k = [1,2,3,4,5]
    ks = []
    for i in k:
        ks.append(stats.kstest(zscore(data,blocksize, i), 'norm'))
        
    result = defaultdict(list)
    for ks in zip(ks):
        result['ks_result'].append(ks)  
    df = pd.DataFrame(data=result)
    df.index = df.index + 1
    filename = 'ks-test-' + str(len(data)) +'-digits-' + str(int(number_of_blocks)) + '-blocks-' + '-digit1-5' 
    df.to_csv(filename + '.csv')   
    
    return ks
        
```

```python
ks_zscore(pi_cf_digits, 10000)
```

```python
stats.kstest(zscore(pi_cf_digits,10000,1), 'norm')
```

```python
stats.kstest(zscore(pi_cf_digits,10000,2), 'norm')
```

```python
stats.kstest(zscore(pi_cf_digits,10000,3), 'norm')
```

# Anderson-Darling for z_score

```python
pi_cf_digits = list(pi_digits['arr_0'])[1:10000001] 
```

```python
def ad_zscore(data, blocksize) :
    number_of_blocks = len(data)/blocksize
    
    k = [1,2,3,4,5]
    ad = []
    for i in k:
        ad.append(stats.anderson(zscore(data,blocksize, i), 'norm'))
        
    result = defaultdict(list)
    for ad in zip(ad):
        result['ad_result'].append(ad)  
    df = pd.DataFrame(data=result)
    df.index = df.index + 1
    filename = 'ad-test-results-' + str(len(data)) +'-digits-' + str(int(number_of_blocks)) + '-blocks-' + '-digit1-5' 
    df.to_csv(filename + '.csv')   
    
    return ks
```

```python
ad_zscore(pi_cf_digits, 10000)
```
