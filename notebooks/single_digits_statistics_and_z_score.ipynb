# Single Digit Statistics and Z-Scores

```python
import numpy as np
from scipy import stats
import math
from collections import defaultdict, Counter
# import matplotlib.pyplot as plt
from tqdm import tqdm
# from scipy.signal import savgol_filter
import pandas as pd
import os  
import time
```

```python
pi_digits = np.load("iteration1_int.npz")
```

```python
pi_cf_digits = list(pi_digits['arr_0'])[1:10000001] # use 10M and blocksize 100k
```

```python
def zscore(data, blocksize=100000, d = 1):

    number_of_blocks = len(data)/blocksize
    
    groups = list(np.array_split(data, number_of_blocks))
    k = []
    for each_list in groups:
        k.append(list(each_list).count(d))
    print(k)
    P = math.log2(1+1/(d*(d+2)))
    z_list = []
    for i in range(10):
        z_list.append((k[i]-blocksize*P)/(math.sqrt(blocksize*P*(1-P))))
    return z_list
```

```python
zscore(pi_cf_digits,100000,1)
```

```python
zscore(pi_cf_digits,100000,2)
```

```python
zscore(pi_cf_digits,100000,3)
```

```python
zscore(pi_cf_digits,100000,4)
```

```python
zscore(pi_cf_digits,100000,5)
```

# Kolmogorov-Smirnov Test for z_score

```python
stats.kstest(zscore(pi_cf_digits,100000,1), 'norm')
```
